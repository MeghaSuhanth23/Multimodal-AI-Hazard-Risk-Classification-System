```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("shlokraval/ppe-dataset-yolov8")

print("Path to dataset files:", path)
```

```python
import os

print("Dataset path:", path)
print("Files inside dataset:")
print(os.listdir(path))
```

```python
image_dir = os.path.join(path, "train", "images")

import cv2
import matplotlib.pyplot as plt

img_path = os.path.join(image_dir, os.listdir(image_dir)[0])

img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.title("Sample PPE Dataset Image")
plt.axis("off")
```

```python
print("Image shape:", img.shape)
label_dir = os.path.join(path, "train", "labels")

label_file = os.listdir(label_dir)[0]

with open(os.path.join(label_dir, label_file), "r") as f:
    print(f.read())
```
